```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('data/sales_data.db')
cursor = conn.cursor()
```

```python
df_1 = pd.read_sql_query("SELECT * FROM top_3_products_by_store", conn)
df_1.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df_1,
    x="branch",
    y="total_sale",
    hue="product_line"
)
plt.title("Top 3 Sales by Product Line for Each Branch", fontsize=16)
plt.ylabel("Total Sale ($) per product line", fontsize=12)
plt.xlabel("Branch", fontsize=12)
plt.legend(title="Product Line", fontsize=10)
plt.tight_layout()
plt.show()
```

```python
df_2 = pd.read_sql_query("SELECT * FROM top_store_by_product", conn)
df_2.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df_2,
    x="product_line",
    y="total_sale",
    hue="branch",  # Highlight the best-performing branch
)
plt.title("Top Store by Product Line", fontsize=16)
plt.ylabel("Total Sale ($)", fontsize=12)
plt.xlabel("Product Line", fontsize=12)
plt.xticks(rotation=30, ha="right")
plt.legend(title="Branch", fontsize=10)
plt.tight_layout()
plt.show()
```
